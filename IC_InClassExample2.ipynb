Importing Data and plotting for population data.


```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Import "Population_Data.csv" as a dataframe called df
df = pd.read_csv("Population_Data.csv")
df.head()
```

```python
# Average the population from days 17 to days 60
avg_population = df[(df['Days'] >= 17) & (df['Days'] <= 60)]['Population'].mean()
print(f"Average population from days 17 to 60: {avg_population}") 
```

```python
#plot the data with days on the x-axis and population on the y-axis
plt.plot(df['Days'], df['Population'], color='black', marker='o', linestyle='-')
plt.axhline(y=276, color='green', linestyle='--', linewidth=2, label='y=276')
plt.xlabel('Days')
plt.ylabel('Population')
plt.title('Population Growth Over Time')
plt.xlim(left=0)
plt.ylim(bottom=0)
plt.grid(True)
plt.legend()
plt.show()
```
