Chapter 5: SIR Models
- Algorithm to predict the fox population after being infected with rabies

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#All the varibles need for an SIR model

# Susceptible population: St = 999
# Infected population: It = 1
# Infecting rate: BETA = .001
# Recovered population: Rt = 0
# Chance of death: d = .5

#S: St+1 = St - BETA*St*It
#I: It+1 = It - d*It+BETA*St*It
#R: Rt+1 = Rt + d*It

#St+1+It+1+Rt+1 = St+It+Rt

#Epidemic Threshold
#BETA*St*It > d*It
```

```python
# Defining the fox population
def fox_population(S0, I0, R0, Beta, d, time):
    current_S0 = S0
    current_I0 = I0
    current_R0 = R0

    # Saves values into a list
    S0_values = [S0] 
    I0_values = [I0]
    R0_values = [R0]
    for _ in range(time):
        current_S0 = current_S0 - (Beta*current_S0*current_I0) # follows the S equation
        current_I0 = current_I0 - (d*current_I0) + (Beta*current_S0*current_I0) # follows the I equation
        current_R0 = current_R0 + (d*current_I0) # follows the R equation

        current_S0 = round(current_S0, 3)
        S0_values.append(current_S0)
        current_I0 = round(current_I0, 3)
        I0_values.append(current_I0)
        current_R0 = round(current_R0, 3)
        R0_values.append(current_R0)

    return S0_values, I0_values, R0_values
```

```python
#Calling the SIR model to run
                            # def fox_population(S0, I0, R0, Beta, d, time)
S0_values, I0_values, R0_values = fox_population(999, 1, 0, .001, .5, 40)

#Printing all the values that were producted from the function
print(S0_values, I0_values, R0_values) 
```

```python
#Graphing the SIR model

# Create time axis for the graph
time_axis = list(range(41)) 

# Plotting the SIR equations onto a graph
plt.figure(figsize=(10, 6))
plt.style.use('dark_background')
plt.plot(time_axis, S0_values, color='blue', label='Susceptible', marker='o')
plt.plot(time_axis, I0_values, color='red', label='Infected', marker='s')
plt.plot(time_axis, R0_values, color='green', label='Removed', marker= '*')
plt.xlabel('Time (Years)')
plt.ylabel('Population')
plt.title('Population of Foxes vs Rabies Over Time')
#plt.xlim(left=0) 
#plt.ylim(bottom=0)
plt.legend()
plt.grid(True)
plt.show()
```

Q1: Beta variable - What happens when it is changed up or down? What are reasons it may change?

    When the Beta variable is change down then it causes an infection to practicaly not exists. On the other hand increasing it cause more of the population to be infected. The reason why these might change has to do with many outside factors like immunity or proximty to susceptible populaiton.

2: d variable - What happens when the death rate is changed? Why might the rate change, or why might it be high/low?

    If the death rate increase the infection has a much lower chance of spreading to other. On the other hand, a lower death rate may cause the infected to linger much longer and spread the infection to the susceptible poplulation. Once again, the rate  might change due to the circumstances of how well a host can combat against the infection or how deadly the infection can be.
