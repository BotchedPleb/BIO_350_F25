```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an algorithm that calculates the population of 3 ages of a beetle population
#  and the total population over 50 years

# the inital first year population is 500 and the inital second year population is 500
# On average 1 beetle produces 1 offsrping in the second year and 5 offspring in the third year
# The first year beetles survive 30% to second year and 50% of second year beetles survive to third year

#Create a definition to calculate and return the population of each age group
# and the total population of each year

def beetle_pop(years,initial_first_year,initial_second_year):
    """
    INPUTS
    ----------
    years (int): The number of years to simulate the beetle population.
    initial_first_year (int): The initial population of first-year beetles.
    initial_second_year (int): The initial population of second-year beetles.

    RETURNS
    ----------
    first_year_population (list): A list containing the population of first-year beetles for each year.
    second_year_population (list): A list containing the population of second-year beetles for each year.
    third_year_population (list): A list containing the population of third-year beetles for each year.
    total_population (list): A list containing the total beetle population for each year.
    """
    
    #Create a matrix to represent the transition of the population between years
    A = np.array([ [0,1,5], [.3,0,0], [0,0.5,0] ])
    initial_population_vector = np.array([initial_first_year,initial_second_year,0])
    #Create an array to store the total population each year
    total_population = []
    first_year_population = []
    second_year_population = []
    third_year_population = []
    #append the current population to the list
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(sum(initial_population_vector))

    n_t = initial_population_vector
    for year in range(years):
        n_t = np.dot(n_t, A)
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(sum(n_t))

    return first_year_population, second_year_population, third_year_population, total_population

```

```python
#Run the function for 50 years with an initial population of 500 first year and 500 second year beetles
first_year_population, second_year_population, third_year_population, total_population = beetle_pop(150, 500, 500)
```

```python
#plot the population of each age group and the total population over time
plt.figure(figsize=(12,8))
plt.plot(first_year_population, label='First Year Population', color='blue')
plt.plot(second_year_population, label='Second Year Population', color='orange')
plt.plot(third_year_population, label='Third Year Population', color='green')
plt.plot(total_population, label='Total Population', linestyle='--', color='black')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time')
plt.legend()
plt.show()
```

```python
# Create an algorithm that calculates the population of 3 ages of a beetle population
#  and the total population over 50 years

# the inital first year population is 500 and the inital second year population is 500
# On average 1 beetle produces 1 offsrping in the second year and 5 offspring in the third year
# The first year beetles survive 30% to second year and 50% of second year beetles survive to third year

#Create a definition to calculate and return the population of each age group
# and the total population of each year

def beetle_pop_withtrap(years,initial_first_year,initial_second_year):
    """
    INPUTS
    ----------
    years (int): The number of years to simulate the beetle population.
    initial_first_year (int): The initial population of first-year beetles.
    initial_second_year (int): The initial population of second-year beetles.

    RETURNS
    ----------
    first_year_population (list): A list containing the population of first-year beetles for each year.
    second_year_population (list): A list containing the population of second-year beetles for each year.
    third_year_population (list): A list containing the population of third-year beetles for each year.
    total_population (list): A list containing the total beetle population for each year.
    """
    
    #Create a matrix to represent the transition of the population between years
    trap_rate = .05
    A = np.array([ [0,1,5], [.3,0,0], [0,0.5,0] ]) * (1-trap_rate)
    initial_population_vector = np.array([initial_first_year,initial_second_year,0])
    #Create an array to store the total population each year

    first_year_population = []
    second_year_population = []
    third_year_population = []
    total_population = []
    
    #append the current population to the list
    first_year_population.append(initial_population_vector[0])
    second_year_population.append(initial_population_vector[1])
    third_year_population.append(initial_population_vector[2])
    total_population.append(sum(initial_population_vector))

    n_t = initial_population_vector
    for year in range(years):
        n_t = np.dot(n_t, A)
        first_year_population.append(n_t[0])
        second_year_population.append(n_t[1])
        third_year_population.append(n_t[2])
        total_population.append(sum(n_t))

    return first_year_population, second_year_population, third_year_population, total_population

```

```python
#Run the function for 50 years with an initial population of 500 first year and 500 second year beetles
first_year_population, second_year_population, third_year_population, total_population = beetle_pop_withtrap(150, 500, 500)

plt.figure(figsize=(12,8))
plt.plot(first_year_population, label='First Year Population', color='blue')
plt.plot(second_year_population, label='Second Year Population', color='orange')
plt.plot(third_year_population, label='Third Year Population', color='green')
plt.plot(total_population, label='Total Population', linestyle='--', color='black')
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('Beetle Population Over Time')
plt.legend()
plt.show()
```
