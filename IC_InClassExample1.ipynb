```python
#import matplotlib library for plotting
#import numpy for numerical operations
import matplotlib.pyplot as plt
import numpy as np
```

```python
#
def N_exp(N_0, p, time):
    # Save N values in a list
    N_values = [N_0] 
    # Current population starts at N_0
    current_N = N_0
    # Iterate for the specified time
    for _ in range(time):
        # Exponential growth formula
        current_N = current_N + p * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        # Append current population to the list
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 40 days and p=0.4 wit a initial population of 0.01
results = N_exp(0.01, 0.4, 35)
print(results)
```

```python
days = list(range(36)) # Create a list of days from 0 to 35
plt.plot(days, results, marker='o')
#adding labels and title
plt.xlabel('Days')
plt.ylabel('Duckweed Population')  
plt.title('Duckweed Population Growth Over Time')
plt.grid(True)

```

```python
days = list(range(36)) # Create a list of days from 0 to 35
plt.plot(days, results, marker='o')
#add a log scale to the y axis
plt.yscale('log')
#adding labels and title
plt.xlabel('Days')
plt.ylabel('Duckweed Population')  
plt.title('Duckweed Population Growth Over Time')
plt.grid(True)
```

```python
#Change N_Exp to N_exp2 and have 3 conditions based on N_0
#if N_0 < 500, p = 0.4
#if N_0 >= 500 and N_0 < 1000, p = 0.2
#if N_0 >= 1000, p = 0

def N_exp2(N_0, time):
    # Save N values in a list
    N_values = [N_0] 
    # Current population starts at N_0
    current_N = N_0
    # Iterate for the specified time
    for _ in range(time):
        # Determine growth rate based on current population
        if current_N < 500:
            p = 0.4
        elif 500 <= current_N < 1000:
            p = 0.2
        else:
            p = 0
        # Exponential growth formula
        current_N = current_N + p * current_N
        # Round to 3 decimal places
        current_N = round(current_N, 3)
        # Append current population to the list
        N_values.append(current_N)
    return N_values
```

```python
#Run the function with 35 days and a initial population of 0.01
result = N_exp2(0.01, 35)
print(result)
```

```python
# Plotting the duckweed population
days = list(range(36)) # Create a list of days from 0 to 35
plt.plot(days, result, marker='o', color='green')
#adding labels and title
plt.xlabel('Days')
plt.ylabel('Duckweed Population')  
plt.title('Duckweed Population Growth Over Time')
plt.grid(True)
```
