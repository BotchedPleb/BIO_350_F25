```python
# Define the population growth model function
def NandM_population(S, dN, dM, N_0, M_0, KN, KM, time):
    current_N = N_0
    current_M = M_0
    N_values = [N_0]
    M_values = [M_0]
    for i in range(time):
        if i == 5:
            current_M = current_M + 100 #give the simulation of a bloom for M plants

        gmN = (S * current_N) * (1 - (current_N / KN) - (current_M / KM)) - dN * current_N
        gmM = (S * current_M) * (1 - (current_M / KM) - (current_N / KN)) - dM * current_M

        current_N += gmN
        current_N = round(current_N, 3)
        N_values.append(current_N)

        current_M += gmM
        current_M = round(current_M, 3)
        M_values.append(current_M)
    return N_values, M_values

# Run the model with an increases in starting population 
Npop, Mpop = NandM_population(1, 0.4, 0.6, 12, 80, 400, 1600, 30)
```

```python
# Create time axis for the graph
time_axis = list(range(31)) 

# Plot the Lotka-Volterra competition equation onto a graph
plt.figure(figsize=(10, 6))
plt.plot(time_axis, Npop, label='Population N', marker='o')
plt.plot(time_axis, Mpop, label='Population M', marker='s')
plt.xlabel('Time (Years)')
plt.ylabel('Population')
plt.title('Population Growth of N and M Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Xn1 = Xt + rx * Xt (1 - ((Xt + alpha * Yt)/Kx))

#rx = maximum population growth rate of X
#kx = carrying capacity
#Competition term a: denotes the per capita
#Effect of Y on X to the effects of X on X.
    ## measure of intensity of interspecific to intraspecific competeitive effects

# Define the population growth model function
def NandM_population(rN, rM, N_0, M_0, KN, KM, time, alpha, beta):
    current_N = N_0
    current_M = M_0
    N_values = [N_0]
    M_values = [M_0]
    for i in range(time):
        if i == 14:
            current_N = current_N + 200 # give the simulation of a bloom for N plants

        current_N = current_N + rN * current_N * (1 - ((current_N + alpha * current_M)/KN)) 
        current_M = current_M + rM * current_M * (1 - ((current_M + beta * current_N)/KM))
        
        current_N = round(current_N, 3)
        N_values.append(current_N)

        current_M = round(current_M, 3)
        M_values.append(current_M)
    return N_values, M_values
```

```python
# Run the model 
       # def NandM_population(rN, rM, N_0, M_0, KN, KM, time, alpha, beta)
Npop, Mpop = NandM_population(1.2, 1.2, 12, 5, 400, 500, 30, 0.5, .9)

# changed the original carrying capacity of 1600 to 500
print(Npop,Mpop)
```

```python
# Create time axis for the graph
time_axis = list(range(31)) 

# Plot the Lotka-Volterra competition equation onto a graph
plt.figure(figsize=(10, 6))
plt.plot(time_axis, Npop, label='Population N', marker='o')
plt.plot(time_axis, Mpop, label='Population M', marker='s')
plt.xlabel('Time (Years)')
plt.ylabel('Population')
plt.title('Population Growth of N and M Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#creates stability when carry capacity is lowered for population M
```
