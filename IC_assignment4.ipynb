Chapter 4: Lokta-Volterra Competition
- Alorgithm to calulate 2 separate plant species to visualize interspecific competition

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#N and M population growth model
#S = 1 viable seed
#dN = death rate of N
#dM = death rate of M
#N_0 = initial population of N
#M_0 = initial population of M
#KN = carrying capacity of N
#KM = carrying capacity of M
#time = number of years to simulate
```

```python
# Define the population growth model function
def NandM_population(S, dN, dM, N_0, M_0, KN, KM, time):
    current_N = N_0
    current_M = M_0
    N_values = [N_0]
    M_values = [M_0]
    for _ in range(time):
        gmN = (S * current_N) * (1 - (current_N / KN) - (current_M / KM)) - dN * current_N
        gmM = (S * current_M) * (1 - (current_M / KM) - (current_N / KN)) - dM * current_M

        current_N += gmN
        current_N = round(current_N, 3)
        N_values.append(current_N)

        current_M += gmM
        current_M = round(current_M, 3)
        M_values.append(current_M)
    return N_values, M_values

# Run the model
Npop, Mpop = NandM_population(1, 0.4, 0.6, 12, 5, 400, 1600, 30)

print(Npop)
print(Mpop)
```

```python
# Create time axis for the graph
time_axis = list(range(31)) 

# Plot the Lotka-Volterra competition equation onto a graph
plt.figure(figsize=(10, 6))
plt.plot(time_axis, Npop, label='Population N', marker='o')
plt.plot(time_axis, Mpop, label='Population M', marker='s')
plt.xlabel('Time (Years)')
plt.ylabel('Population')
plt.title('Population Growth of N and M Over Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
