Plot the data showing the frequency of your data with the probability density on the y axis

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Simulate Poisson-distributed data: number of customer arrivals per hour
np.random.seed(1)
lam = 6  # average rate (lambda)
data = np.random.poisson(lam=lam, size=1000) 

# Plot frequency with probability density
plt.figure(figsize=(8,5))
sns.histplot(data, bins=range(0, max(data)+2), stat='density', 
             color='orange', edgecolor='black', alpha=0.7)

# Overlay theoretical Poisson distribution
x = np.arange(0, max(data)+1)
plt.plot(x, poisson.pmf(x, lam), 'r-', lw=2)

plt.title('Poisson Distribution: Customer Arrivals per Hour')
plt.xlabel('Number of Customers')
plt.ylabel('Probability Density')
plt.grid(alpha=0.3)
plt.show()
```
